```python
import pandas as pd
import os
```

```python
dfs = []
for file in os.listdir('Eye-tracking Output/'):
    dfs.append(pd.read_csv('Eye-tracking Output/' + file))
dfs[24]
```

```python
instances = pd.concat([df[['Participant', 'Stimulus']].drop_duplicates() for df in dfs], ignore_index=True)
instances
```

```python
chopped = []
for df in dfs:
    grouped = df.groupby(['Participant', 'Stimulus'])
    for name, group in grouped:
        chopped.append(group)
```

```python
output='split'
if output not in os.listdir(): os.mkdir(output)
for i, exp in enumerate(chopped):
    exp.to_csv(os.path.join(output, str(i) + '.csv'))
```
